### **Outliers**

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Pada data ini terlihat pada kolom:
* Harga per unit = 0% outlier
* Unit terjual = 0% outlier
* Total pendapatan per transaksi = 0,4% outlier

##**Distribution/Distribusi**

Aktivitas:
Menganalisis distribusi data pada variabel numerik yaitu Price_USD, Units_Sold, dan Revenue_USD.

Tujuan:
Untuk memahami pola penyebaran data, mengidentifikasi kecenderungan distribusi (skewness), serta mendeteksi adanya outlier pada masing-masing variabel.

Visualisasi:
Histogram

```python
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns

for col in numerical_cols:
    plt.figure(figsize=(10, 6))
    sns.histplot(df[col], bins=20, kde=True)
    plt.title(f'Distribusi {col}')
    plt.xlabel(col)
    plt.ylabel('Frekuensi')
    plt.show()
```

Ketiga histogram ini secara bersamaan memberikan gambaran tentang karakteristik data penjualan sepatu dari sisi harga, kuantitas, dan pendapatan yang dihasilkan. Menariknya, ketiga variabel ini menunjukkan pola distribusi yang berbeda-beda, dan justru di situlah letak insight yang paling berharga.

* Price_USD memiliki distribusi yang relatif datar dan merata sepanjang rentang $20 hingga $250. Hampir semua bin frekuensinya berkisar di angka 40-58, tanpa ada satu rentang harga yang benar-benar mendominasi. Kurvanya pun bergelombang tanpa puncak tunggal yang jelas, menandakan bahwa produk yang dijual mencakup berbagai segmen harga secara seimbang.

* Units_Sold menunjukkan pola yang sangat serupa dengan Price_USD, yakni distribusi seragam (uniform) dengan frekuensi yang konsisten di kisaran 40-60 sepanjang rentang 1 hingga 20 unit. Tidak ada kuantitas pembelian tertentu yang jauh lebih dominan dari yang lain, yang berarti perilaku pembelian konsumen sangat bervariasi tanpa pola yang terpusat.

* Revenue_USD justru menunjukkan pola yang sama sekali berbeda, yaitu distribusi miring ke kanan (right-skewed) yang sangat kentara. Puncak frekuensi tertinggi berada di rentang $200-$600 dengan frekuensi mencapai 132, lalu turun secara konsisten seiring meningkatnya nilai pendapatan. Transaksi di atas $2.500 sudah sangat jarang, dan di atas $4.000 hampir tidak ada.

Di sinilah interpretasi menjadi lebih menarik. Meski harga (Price_USD) dan kuantitas (Units_Sold) keduanya terdistribusi merata, pendapatan (Revenue_USD) yang merupakan hasil perkalian keduanya justru terpusat di nilai rendah. Ini mengindikasikan bahwa kombinasi harga tinggi dan kuantitas banyak dalam satu transaksi adalah kejadian yang jarang, sehingga sebagian besar pendapatan yang terbentuk tetap berada di kisaran yang lebih kecil. Dengan kata lain, transaksi yang menghasilkan revenue besar membutuhkan kondisi yang tidak selalu terpenuhi secara bersamaan harga produk yang mahal sekaligus dibeli dalam jumlah banyak.

Ketiga distribusi ini secara bersama-sama menggambarkan sebuah bisnis yang melayani pasar yang sangat luas dan heterogen, baik dari sisi segmen harga maupun kebiasaan beli konsumennya. Namun demikian, pendapatan yang dihasilkan tetap cenderung didominasi oleh transaksi-transaksi bernilai kecil hingga menengah. Strategi yang paling relevan dari temuan ini adalah mendorong peningkatan frekuensi transaksi secara masif, karena mengandalkan transaksi bernilai besar saja tidak cukup untuk menjaga stabilitas pendapatan bisnis.

##**Relationship/Hubungan**

Aktivitas: Menganalisis tingkat kekuatan korelasi atau hubungan antara jumlah donasi yang disetujui dengan jumlah donasi yang disalurkan.

Tujuan: Untuk mengetahui seberapa konsisten proses penyaluran dana dibandingkan dengan jumlah dana yang telah disetujui, serta mengidentifikasi apakah terdapat keterikatan yang kuat di antara kedua variabel finansial tersebut.

Visualisasi: Heatmap

```python
plt.figure(figsize=(12, 10))
sns.heatmap(data=df.select_dtypes(include=['float64', 'int64']).corr(),
            annot=True,
            cmap='viridis',
            fmt='.2f')
plt.title('Hubungan Korelasi Antar Semua Kolom Numerik')
plt.show()
```

Heatmap korelasi ini menampilkan kekuatan hubungan antar tiga variabel numerik, yaitu Price_USD, Units_Sold, dan Revenue_USD. Nilai korelasi berkisar dari -1 hingga 1, di mana semakin mendekati 1 berarti hubungannya semakin kuat dan positif, semakin mendekati -1 berarti kuat namun berlawanan arah, dan mendekati 0 berarti hampir tidak ada hubungan linear.

* Price_USD vs Units_Sold — Korelasi: -0,02
Nilai ini praktis mendekati nol, artinya hampir tidak ada hubungan linear antara harga produk dan jumlah unit yang terjual. Konsumen tidak secara konsisten membeli lebih sedikit ketika harga naik, maupun membeli lebih banyak ketika harga turun. Ini mengindikasikan bahwa keputusan kuantitas pembelian tidak dipengaruhi oleh harga produk secara signifikan dalam dataset ini.

* Price_USD vs Revenue_USD — Korelasi: 0,59
Terdapat korelasi positif yang cukup kuat antara harga dan pendapatan. Artinya, produk dengan harga yang lebih tinggi cenderung menghasilkan pendapatan yang lebih besar, yang memang masuk akal secara logis. Namun nilai 0,59 yang tidak mencapai 1,00 menunjukkan bahwa harga bukan satu-satunya penentu revenue faktor kuantitas juga turut berperan.

* Units_Sold vs Revenue_USD — Korelasi: 0,73
Ini adalah korelasi tertinggi di antara semua pasangan variabel. Hubungan antara jumlah unit terjual dan pendapatan terbilang kuat, yang berarti semakin banyak unit yang terjual, semakin besar pendapatan yang dihasilkan. Korelasi yang lebih tinggi dibanding Price_USD vs Revenue_USD menunjukkan bahwa volume penjualan memiliki pengaruh yang lebih dominan terhadap pendapatan dibandingkan harga satuan produk.

Kesimpulan
Temuan paling krusial dari heatmap ini adalah bahwa volume penjualan (Units_Sold) lebih berpengaruh terhadap pendapatan dibandingkan harga produk (Price_USD). Sementara itu, harga sama sekali tidak memengaruhi berapa banyak unit yang dibeli konsumen. Implikasi strategisnya cukup jelas jika tujuannya adalah memaksimalkan Revenue_USD, maka mendorong peningkatan volume penjualan harus menjadi prioritas utama, diikuti oleh optimalisasi harga sebagai faktor pendukung.
